## **Metodología Empleada de Scraping**

Para el proceso de recolección de datos, utilizaremos **Selenium**, una herramienta de automatización de navegadores ampliamente utilizada en proyectos de web scraping que requieren interactuar con sitios web dinámicos. **Selenium** nos permitirá simular la navegación en la **plataforma de Mercado Libre**, posibilitando la cargar dinámica del contenido de la página y el manejo de interacciones complejas como el desplazamiento y los clics.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from tabulate import tabulate
import json

# Configuración del navegador
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=options)
url = "https://listado.mercadolibre.com.co/carro-control-remoto"
driver.get(url)

# Esperar a que la página cargue dinámicamente
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'li.ui-search-layout__item')))

# Buscar todos los productos
productos = driver.find_elements(By.CSS_SELECTOR, 'li.ui-search-layout__item')

# Lista para almacenar los datos
resultado_productos = []

print("\n=== PRODUCTOS ENCONTRADOS ===\n")

for idx, producto in enumerate(productos, start=1):
    try:
        # Extraer información básica
        titulo = producto.find_element(By.TAG_NAME, 'h2').text
        precio_elemento = producto.find_element(By.CSS_SELECTOR, "span.andes-money-amount")
        simbolo = precio_elemento.find_element(By.CSS_SELECTOR, "span.andes-money-amount__currency-symbol").text
        fraccion = precio_elemento.find_element(By.CSS_SELECTOR, "span.andes-money-amount__fraction").text
        precio = f"{simbolo}{fraccion}"
        link = producto.find_element(By.TAG_NAME, "a").get_attribute("href")

        # Obtener calificación
        try:
            calificacion = producto.find_element(By.CSS_SELECTOR, "span.poly-reviews__rating").text
        except Exception:
            calificacion = "No disponible"

        # Obtener tipo de envío
        try:
            tipo_envio = producto.find_element(By.CSS_SELECTOR, "div.poly-component__shipping").text
        except Exception:
            tipo_envio = "No disponible"

        # Verificar disponibilidad del producto
        try:
            disponibilidad = producto.find_element(By.CSS_SELECTOR, "span.ui-search-item__availability").text
        except Exception:
            disponibilidad = "En stock"  # Asumir en stock si no se encuentra el elemento

        # Añadir a la lista de resultados
        resultado_productos.append({
            'Título': titulo,
            'Precio': precio,
            'Enlace': link,
            'Calificación': calificacion,
            'Tipo de envío': tipo_envio,
            'Disponibilidad': disponibilidad
        })

    except Exception as e:
        print(f"Error al procesar el producto #{idx}: {e}\n")

# Mostrar los resultados en una tabla
tabla_resultados = [
    [prod['Título'], prod['Precio'], prod['Calificación'], prod['Tipo de envío'], prod['Disponibilidad'], prod['Enlace']]
    for prod in resultado_productos
]
encabezados = ["Título", "Precio", "Calificación", "Tipo de Envío", "Disponibilidad", "Enlace"]
print(tabulate(tabla_resultados, headers=encabezados, tablefmt="fancy_grid"))

# Guardar los resultados en un archivo JSON
with open('productos_mejorados.json', 'w', encoding='utf-8') as f:
    json.dump(resultado_productos, f, ensure_ascii=False, indent=4)

print(f"\nTotal de productos encontrados: {len(resultado_productos)}")
print("Los datos han sido guardados en 'productos_mejorados.json'")

# Cerrar el navegador
driver.quit()
```

### **Mejoras Realizadas al Código**

**Manejo de Disponibilidad:**

Si no se encuentra información explícita sobre disponibilidad, el valor por defecto es "En stock".

**Formato Tabular para Resultados:**

Los datos de los productos se presentan en una tabla clara y organizada en la consola, utilizando la biblioteca tabulate.

**JSON Ajustado:**

Los datos se almacenan en un archivo productos_mejorados.json con los campos ajustados: Título, Precio, Enlace, Calificación, Tipo de Envío, Disponibilidad.

**Robustez:**

El manejo de excepciones asegura que el código no falle si algún dato no está disponible para uno o más productos.

Codigo mejorado para el rankeo de productos

```python
#Listado con rankeo con productos de menor a mayor precio
from selenium import webdriver
from selenium.webdriver.common.by import By
from tabulate import tabulate
import time
import json

def setup_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    return webdriver.Chrome(options=options)

def extract_product_info(producto):
    # Información básica
    titulo = producto.find_element(By.TAG_NAME, 'h2').text
    precio_element = producto.find_element(By.CSS_SELECTOR, "span.andes-money-amount")
    simbolo = precio_element.find_element(By.CSS_SELECTOR, "span.andes-money-amount__currency-symbol").text
    fraccion = precio_element.find_element(By.CSS_SELECTOR, "span.andes-money-amount__fraction").text
    precio = float(fraccion.replace(',', ''))  # Convertir a float para ordenar
    link = producto.find_element(By.TAG_NAME, "a").get_attribute("href")

    # Calificación
    try:
        calificacion = producto.find_element(By.CSS_SELECTOR, "span.poly-reviews__rating").text
    except Exception:
        calificacion = "No disponible"

    # Tipo de envío
    try:
        tipo_envio = producto.find_element(By.CSS_SELECTOR, "div.poly-component__shipping").text
    except Exception:
        tipo_envio = "No disponible"

    # Disponibilidad
    try:
        disponibilidad = producto.find_element(By.CSS_SELECTOR, "span.ui-search-item__availability").text
    except Exception:
        disponibilidad = "En stock"  # Asumir en stock si no se encuentra el elemento

    return {
        'titulo': titulo,
        'precio': precio,  # Guardamos como float para ordenar
        'enlace': link,
        'precio_formateado': f"{simbolo}{fraccion}",  # Para impresión
        'calificacion': calificacion,
        'tipo_envio': tipo_envio,
        'disponibilidad': disponibilidad
    }

def main():
    driver = setup_driver()
    url = "https://listado.mercadolibre.com.co/carro-control-remoto"
    driver.get(url)

    # Esperar a que la página cargue
    time.sleep(5)

    productos = driver.find_elements(By.CSS_SELECTOR, 'li.ui-search-layout__item')
    resultado_productos = []

    print("\n=== PRODUCTOS RANKEADOS DE MENOR A MAYOR PRECIO ===\n")

    for producto in productos:
        try:
            producto_info = extract_product_info(producto)
            resultado_productos.append(producto_info)
        except Exception as e:
            print(f"Error al obtener datos de un producto: {e}")

    # Ordenar por precio, del más económico al más costoso
    resultado_productos.sort(key=lambda x: x['precio'])

    # Mostrar resultados en tabla con posición numérica
    tabla_resultados = [
        [idx + 1, prod['titulo'], prod['precio_formateado'], prod['calificacion'], prod['tipo_envio'], prod['disponibilidad'], prod['enlace']]
        for idx, prod in enumerate(resultado_productos)
    ]
    encabezados = ["#", "Título", "Precio", "Calificación", "Tipo de Envío", "Disponibilidad","Enlace"]
    print(tabulate(tabla_resultados, headers=encabezados, tablefmt="fancy_grid"))

    # Guardar los resultados ordenados en un archivo JSON
    with open('productos_ordenados.json', 'w', encoding='utf-8') as f:
        json.dump(resultado_productos, f, ensure_ascii=False, indent=4)

    driver.quit()

if __name__ == "__main__":
    main()

```

### **Mejoras Realizadas**

**Disponibilidad del Producto:**

Se agregó un campo que indica si el producto está disponible o no. Si no se encuentra información, se asume "En stock".

**Calificación:**

Se extrae información de la calificación otorgada por los usuarios al producto. Si no está disponible, se muestra "No disponible".

**Tipo de Envío:**

Se extrae información sobre el tipo de envío. Si no está disponible, se muestra "No disponible".

**Tabla con Posición Numérica:**

Los resultados se imprimen en una tabla organizada con una nueva columna que indica el número de posición según el orden de menor a mayor precio.

**JSON Ajustado:**

Se añadió información de disponibilidad y tipo de envío al archivo JSON exportado.

**Formato Más Claro:**

La consola muestra los productos en una tabla fácil de leer gracias a la biblioteca tabulate.
