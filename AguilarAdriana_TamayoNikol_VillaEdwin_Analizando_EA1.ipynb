## **Metodología Empleada de Scraping**

Para el proceso de recolección de datos, utilizaremos **Selenium**, una herramienta de automatización de navegadores ampliamente utilizada en proyectos de web scraping que requieren interactuar con sitios web dinámicos. **Selenium** nos permitirá simular la navegación en la **plataforma de Mercado Libre**, posibilitando la cargar dinámica del contenido de la página y el manejo de interacciones complejas como el desplazamiento y los clics.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import json

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=options)
url = "https://listado.mercadolibre.com.co/carro-control-remoto"
driver.get(url)

# Esperar a que la página cargue
time.sleep(5)

# Buscar todos los productos
productos = driver.find_elements(By.CSS_SELECTOR, 'li.ui-search-layout__item')

# Lista para almacenar los datos
resultado_productos = []

# Contador para enumerar los productos
contador = 1

print("\n=== PRODUCTOS ENCONTRADOS ===\n")

for producto in productos:
    try:
        # Extraer información
        titulo = producto.find_element(By.TAG_NAME, 'h2').text
        precio_elemento = producto.find_element(By.CSS_SELECTOR, "span.andes-money-amount")
        simbolo = precio_elemento.find_element(By.CSS_SELECTOR, "span.andes-money-amount__currency-symbol").text
        fraccion = precio_elemento.find_element(By.CSS_SELECTOR, "span.andes-money-amount__fraction").text
        precio = f"{simbolo}{fraccion}"
        link = producto.find_element(By.TAG_NAME, "a").get_attribute("href")

        # Obtener la calificación, si no existe mostrar No disponible
        try:
            calificacion = producto.find_element(By.CSS_SELECTOR, "span.poly-reviews__rating").text
        except Exception:
            calificacion = "No disponible"

        # Imprimir información de manera organizada
        print(f"Producto #{contador}")
        print("-" * 50)
        print(f"Título: {titulo}")
        print(f"Precio: {precio}")
        print(f"Enlace: {link}")
        print(f"Calificación: {calificacion}")
        print("-" * 50 + "\n")

        # Incrementar contador
        contador += 1

        # Añadir a la lista de resultados
        resultado_productos.append({
            'titulo': titulo,
            'precio': precio,
            'enlace_imagen': link,
            'calificacion': calificacion
        })

    except Exception as e:
        print(f"Error al obtener datos del producto #{contador}: {e}\n")

# Guardar los resultados en un archivo JSON
with open('productos.json', 'w', encoding='utf-8') as f:
    json.dump(resultado_productos, f, ensure_ascii=False, indent=4)

print(f"\nTotal de productos encontrados: {contador - 1}")
print("Los datos han sido guardados en 'productos.json'")

# Cerrar el navegador
driver.quit()
```

```python
#Listado con rankeo con productos de menor a mayor precio

from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import json

def setup_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    return webdriver.Chrome(options=options)

def extract_product_info(producto):
    titulo = producto.find_element(By.TAG_NAME, 'h2').text
    precio_element = producto.find_element(By.CSS_SELECTOR, "span.andes-money-amount")
    simbolo = precio_element.find_element(By.CSS_SELECTOR, "span.andes-money-amount__currency-symbol").text
    fraccion = precio_element.find_element(By.CSS_SELECTOR, "span.andes-money-amount__fraction").text
    precio = float(fraccion.replace(',', ''))  # Convertir a float para ordenar, asumiendo que no hay centavos
    link = producto.find_element(By.TAG_NAME, "a").get_attribute("href")
    # Obtener la calificación, si no existe mostrar No disponible
    try:
        calificacion = producto.find_element(By.CSS_SELECTOR, "span.poly-reviews__rating").text
    except Exception:
        calificacion = "No disponible"
    return {
        'titulo': titulo,
        'precio': precio,  # Guardamos como float para ordenar
        'enlace_imagen': link,
        'precio_formateado': f"{simbolo}{fraccion}",  # Para impresión
        'calificacion': calificacion
    }

def main():
    driver = setup_driver()
    url = "https://listado.mercadolibre.com.co/carro-control-remoto"
    driver.get(url)

    # Esperar a que la página cargue
    time.sleep(5)

    productos = driver.find_elements(By.CSS_SELECTOR, 'li.ui-search-layout__item')
    resultado_productos = []

    print("\n=== PRODUCTOS RANKEADOS DE MENOR A MAYOR PRECIO===\n")

    for producto in productos:
        try:
            producto_info = extract_product_info(producto)
            resultado_productos.append(producto_info)
        except Exception as e:
            print(f"Error al obtener datos de un producto: {e}")

    # Ordenar por precio, del más económico al más costoso
    resultado_productos.sort(key=lambda x: x['precio'])

    for producto in resultado_productos:

        print(f"Título: {producto['titulo']}")
        print(f"Precio: {producto['precio_formateado']}")
        print(f"Enlace Imagen: {producto['enlace_imagen']}")
        print(f"Calificación: {producto['calificacion']}")
        print("-" * 50)

    # Guardar los resultados ordenados en un archivo JSON
    with open('productos_ordenados.json', 'w', encoding='utf-8') as f:
        json.dump(resultado_productos, f, ensure_ascii=False, indent=4)

    driver.quit()

if __name__ == "__main__":
    main()
```
